```python
fdir = '/home/users/liujx/bridge_lamb3/das_bhm/'
psds = np.load(f"{fdir}data/fig5a.npz")
das_psd = psds['das_psd']
acc_psd = psds['acc_psd']
ldv_psd = psds['ldv_psd']
das_f = psds['das_f']
acc_f = psds['acc_f']
ldv_f = psds['ldv_f']
```

```python
data = np.load(f"{fdir}data/fig5_ms.npz")
modes_das = data['modes_das']
modes_opensees = data['modes_opensees']
dist_das = data['dist_das']
dist_opensees = data['dist_opensees']
```

```python
for modenum in range(3):
    fig,ax = plt.subplots(figsize=(9,4))
    plt.plot(dist_das,modes_das[modenum],'ko--')
    plt.plot(dist_opensees,modes_opensees[modenum],'r-')  
    plt.xlabel('Distance (m)',fontsize=24)
    plt.ylabel('Normalized \nmode shape',fontsize=24)
    ax.tick_params(axis='both', which='major', labelsize=24)
    if modenum==0:
        plt.legend(['DAS','Ref'],fontsize=20)
    plt.tight_layout()
    plt.grid()
    plt.savefig(f'{fdir}figs/fig5{modenum+1}.pdf')
    print(compute_mac(modes_das[modenum],dist_das,modes_opensees[modenum],dist_opensees))
```
