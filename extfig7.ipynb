```python
import numpy as np
import matplotlib.pyplot as plt
from utils import *
```

```python
fdir = '/home/users/liujx/bridge_lamb3/das_bhm/'
psds = np.load(f"{fdir}data/extfig7a.npz")
das_psd = psds['das_psd']
acc_psd = psds['acc_psd']
das_f = psds['das_f']
acc_f = psds['acc_f']
```
