```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fdir = '/home/users/liujx/bridge_lamb3/das_bhm/'
data = np.load(f"{fdir}data/extfig4.npz")
t_das = data['t_das']
t = data['t']
das = data['das']
ldv = data['ldv']
dis = data['dis']
```

```python
fig,ax = plt.subplots(figsize=(9,5))
ax.plot(t_das, das, color="black", label="DAS")
ax.plot(t, ldv, color="cornflowerblue", label="LDV")
ax.plot(t, dis, color="red", label="DIS")
plt.xlabel('Time (s)',fontsize=24)
plt.ylabel("Displacement (mm)",fontsize=24)
plt.xlim([0,300])
plt.legend(['DAS','LDV','DIS'],fontsize=20, loc='lower right')
ax.tick_params(axis='both', which='major', labelsize=24)
plt.tight_layout()
plt.savefig(f'{fdir}figs/extfig4b.pdf')
print(np.sqrt(np.mean((ldv - das) ** 2)))
```
